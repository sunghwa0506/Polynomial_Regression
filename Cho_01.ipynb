```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
#a 
#Generate 20 data pairs (X, Y) using y = sin(2*pi*X) + N
#Use uniform distribution between 0 and 1 for X
# Sample N from the normal gaussian distribution
# Use 10 for train and 10 for test

m=20
x=np.random.uniform(0,1,size = (m,1))
n=np.random.uniform(size=(m,1))
y=np.sin(2 * np.pi * x)+n

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5)


plt.scatter(x_train,y_train,color='r')
plt.scatter(x_test,y_test,color='b')
plt.show
```

```python


x_train=x_train.reshape(1,-1)
x_test=x_test.reshape(1,-1)

x_train=x_train[0]
x_test=x_test[0]

y_train=y_train.reshape(1,-1)
y_test=y_test.reshape(1,-1)

y_train=y_train[0]
y_test=y_test[0]

data=pd.DataFrame({"x":x_train,
                  "y":y_train})
data_test=pd.DataFrame({"x_test":x_test,
                  "y_test":y_test})


data=data.sort_values('x')
data_test=data_test.sort_values('x_test')

x_train=data['x'].tolist()
x_test=data_test['x_test'].tolist()

y_train=data['y'].tolist()
y_test=data_test['y_test'].tolist()


x_train = np.reshape(x_train,(-1,1))
x_test = np.reshape(x_test,(-1,1))

y_train = np.reshape(y_train,(-1,1))
y_test = np.reshape(y_test,(-1,1))

y_original=np.sin(2 * np.pi * x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
rmse_train_poly_0 = (np.sqrt(mean_squared_error(y_train,y_train_poly_predict_0)))
rmse_test_poly_0 = (np.sqrt(mean_squared_error(y_test,y_test_poly_predict_0)))
```
